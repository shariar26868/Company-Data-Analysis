# Data Analysis

```python
import pandas as pd
new_data_frame=pd.read_csv('monthwise_data/final.csv')
new_data_frame
```

```python
dataset=new_data_frame.copy()
```

# Requirement_1- Compare the performance of different buyers or styles in terms of achieved working hours.

```python
buyer_performance = dataset.groupby('Buyer ')['Wr- Hr'].sum().reset_index()
style_performance = dataset.groupby('Style No')['Wr- Hr'].sum().reset_index()
buyer_performance = buyer_performance.sort_values(by='Wr- Hr', ascending=False)
style_performance = style_performance.sort_values(by='Wr- Hr', ascending=False)

# Print the results
print("Buyer Performance:")
print(buyer_performance)

print("\nStyle Performance:")
print(style_performance)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(buyer_performance['Buyer '], buyer_performance['Wr- Hr'], color='blue')
plt.xlabel('Buyer ')
plt.ylabel('Total Working Hours')
plt.title('Buyer Performance - Total Working Hours')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(style_performance['Style No'], style_performance['Wr- Hr'], color='green')
plt.xlabel('Style No')
plt.ylabel('Total Working Hours')
plt.title('Style Performance - Total Working Hours')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Requirement_2- How well does the actual manpower match the planned manpower in terms of meeting the target hour?

```python
dataset['Manpower Efficiency'] = dataset['Actual Man Power HP'] + dataset['Actual Man Power OP'] - \
                                 dataset['Plan Man Power. HP'] - dataset['Plan Man Power. OP']
plt.figure(figsize=(10, 6))
plt.scatter(dataset['Manpower Efficiency'], dataset['Target Hour.'], color='red', alpha=0.5)
plt.xlabel('Manpower Efficiency (Actual - Planned)')
plt.ylabel('Target Hour')
plt.title('Relationship Between Manpower Efficiency and Target Hour')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Requirement_3- How do target hours change over different periods?

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
periodly_target_hours = dataset.groupby(dataset['Date'].dt.to_period("M"))['Target Hour.'].mean().reset_index()
periodly_target_hours['Date'] = periodly_target_hours['Date'].astype(str)
plt.figure(figsize=(12, 6))
plt.plot(periodly_target_hours['Date'], periodly_target_hours['Target Hour.'], marker='o', linestyle='-', color='b')
plt.xlabel('Period')
plt.ylabel('Average Target Hours')
plt.title('Change in Target Hours Over Different Periods')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```
